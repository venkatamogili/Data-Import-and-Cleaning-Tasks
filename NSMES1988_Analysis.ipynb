## 8. Memory Analysis and Optimization

Analyze the current memory usage and recommend optimized data types for memory efficiency.

```python
# Memory Analysis and Optimization Recommendations
print("=" * 80)
print("MEMORY ANALYSIS AND OPTIMIZATION")
print("=" * 80)

# Current memory usage
current_memory = df.memory_usage(deep=True).sum() / 1024**2
print(f"\n1. CURRENT MEMORY USAGE:")
print(f"   Total: {current_memory:.2f} MB")
print(f"\n   Memory by Column (KB):")
print(df.memory_usage(deep=True) / 1024)

# Show current dtypes
print(f"\n2. CURRENT DATA TYPES:")
print(df.dtypes)

# Recommendations for optimization
print("\n" + "=" * 80)
print("OPTIMIZATION RECOMMENDATIONS")
print("=" * 80)

recommendations = {
    'Unnamed: 0': 'Drop - index column (redundant)',
    'visits': 'int16 (currently int64) - values are small integers',
    'nvisits': 'int8 (currently int64) - values are small integers',
    'ovisits': 'int8 (currently int64) - values are small integers',
    'novisits': 'int8 (currently int64) - values are small integers',
    'emergency': 'int8 (currently int64) - values are small integers',
    'hospital': 'int8 (currently int64) - values are small integers',
    'chronic': 'int8 (currently int64) - values are small integers',
    'school': 'int8 (currently int64) - values are small integers',
    'health': 'category - limited unique values',
    'adl': 'category - limited unique values',
    'region': 'category - limited unique values',
    'gender': 'category - binary categorical',
    'married': 'category - binary categorical',
    'employed': 'category - binary categorical',
    'insurance': 'category - binary categorical',
    'medicaid': 'category - binary categorical',
    'age': 'float32 (currently float64) - precision loss acceptable',
    'income': 'float32 (currently float64) - precision loss acceptable'
}

print("\nRECOMMENDED OPTIMIZATIONS:")
for col, rec in recommendations.items():
    print(f"   {col}: {rec}")

# Calculate potential savings
print("\n" + "=" * 80)
print("POTENTIAL MEMORY SAVINGS CALCULATION")
print("=" * 80)

df_optimized = df.copy()

# Apply optimizations
df_optimized = df_optimized.drop('Unnamed: 0', axis=1)

# Convert integers
int_columns = ['visits', 'nvisits', 'ovisits', 'novisits', 'emergency', 'hospital', 'chronic', 'school']
for col in int_columns:
    if col in df_optimized.columns:
        df_optimized[col] = df_optimized[col].astype('int8')

# Convert floats
df_optimized['age'] = df_optimized['age'].astype('float32')
df_optimized['income'] = df_optimized['income'].astype('float32')

# Convert to category
cat_columns = ['health', 'adl', 'region', 'gender', 'married', 'employed', 'insurance', 'medicaid']
for col in cat_columns:
    if col in df_optimized.columns:
        df_optimized[col] = df_optimized[col].astype('category')

optimized_memory = df_optimized.memory_usage(deep=True).sum() / 1024**2
original_memory = df.memory_usage(deep=True).sum() / 1024**2
memory_saved = original_memory - optimized_memory
savings_percent = (memory_saved / original_memory) * 100

print(f"\nOriginal Memory Usage: {original_memory:.2f} MB")
print(f"Optimized Memory Usage: {optimized_memory:.2f} MB")
print(f"Memory Saved: {memory_saved:.2f} MB ({savings_percent:.1f}%)")
print(f"\nâœ“ Optimization successful! Data can be compressed by {savings_percent:.1f}%")

# Show optimized dtypes
print(f"\n4. OPTIMIZED DATA TYPES:")
print(df_optimized.dtypes)
```

```python
# --- Load NSMES1988updated.csv into a working DataFrame (`df_work`) ---
import os
import pandas as pd
path = 'NSMES1988updated.csv'
if not os.path.exists(path):
    raise FileNotFoundError(f'{path} not found in the notebook directory')
df_work = pd.read_csv(path)
print('Loaded', path, '-> df_work shape:', df_work.shape)

# Data types and quick info
print("\nData types:")
print(df_work.dtypes)
print("\nDataFrame info:")
df_work.info()

# Identify categorical columns (object/category and low-cardinality numeric)
cat_candidates = df_work.select_dtypes(include=['object','category']).columns.tolist()
low_card = [col for col in df_work.columns if df_work[col].nunique(dropna=True) <= 50 and col not in cat_candidates]
categorical = sorted(set(cat_candidates + low_card))
print("\nCategorical candidate columns:")
print(categorical)
# Ensure `health` and `region` are treated as categorical for analysis
for c in ['health','region']:
    if c in df_work.columns:
        df_work[c] = df_work[c].astype('category')
    else:
        print(f'Warning: {c} not found in df_work columns')

print("\nConfirmed categorical dtypes:")
print(df_work.dtypes[df_work.dtypes == 'category'])

# --- Pivot tables grouped by health (rows) and region (columns) ---
print("\nComputing pivot tables (counts, mean visits, mean income, mean age) grouped by health x region")
pivot_count = pd.pivot_table(df_work, index='health', columns='region', values='visits', aggfunc='count', fill_value=0)
pivot_mean_visits = pd.pivot_table(df_work, index='health', columns='region', values='visits', aggfunc='mean')
pivot_mean_income = pd.pivot_table(df_work, index='health', columns='region', values='income', aggfunc='mean')
pivot_mean_age = pd.pivot_table(df_work, index='health', columns='region', values='age', aggfunc='mean')

print("\nPivot - counts (visits):")
print(pivot_count)

print("\nPivot - mean visits:")
print(pivot_mean_visits)

print("\nPivot - mean income:")
print(pivot_mean_income)

print("\nPivot - mean age:")
print(pivot_mean_age)

# Save working dataframe for future use
df_work.to_csv('NSMES1988_analysis_df.csv', index=False)
print('\nSaved analysis dataframe to NSMES1988_analysis_df.csv')
```
